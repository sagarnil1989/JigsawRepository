```python
## Kmeans, k=24
data_clustering=data.drop(['Player Id','Last_Name','First_Name'],axis=1)
```

```python
data_clustering.head()
```

```python
import sklearn.preprocessing as preprocessing
```

```python
scalers=preprocessing.StandardScaler()
```

```python
data_scaled=scalers.fit_transform(data_clustering)
```

```python
data_scaled
```

```python
### How do we use KMeans sklearn?
import sklearn.cluster as cluster
```

```python
mod=cluster.KMeans(n_clusters=24)
```

```python
mod.fit(data_scaled)
```

```python
mod.labels_
```

```python
data.head()
```

```python
data['cluster_label']=mod.labels_
```
